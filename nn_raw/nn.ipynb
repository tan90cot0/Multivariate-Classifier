Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow

%matplotlib inline  
```

Loading Dataset and filling NaN with 0

```python
filename = "data/original_data.csv"
csv = pd.read_csv(filename)
csv = csv.drop(['Reference'], axis=1)
csv = csv.fillna(0)
csv.head()
```

Extracting features and Labels

```python
m = csv.shape[0]
labels = csv['Actuator Type'].to_numpy().reshape((m,1))


X = csv.drop(['Actuator Type'], axis=1).to_numpy()

# print(labels)
classes = {'PZT': 0.0, 'DEA': 1.0, 'IPMC': 2.0, 'SMA': 3.0, 'SFA': 4.0, 'TSA': 5.0, 'SCP': 6.0, 'EAP': 7.0, 'SMP': 8.0}

for i in range(m):
    labels[i,0] = classes[labels[i,0]]
labels = labels.astype(np.float32)
print(labels)
```

```python
labels = labels.ravel()
print(labels.shape)
# xx = X[:,:2]
print(X.shape)
```

NN model definition, compilation and training with Adam Optimizer

```python
model = tensorflow.keras.Sequential([
    tensorflow.keras.layers.Flatten(input_shape=(5,1)),
    tensorflow.keras.layers.Dense(25, activation='relu'),
    tensorflow.keras.layers.Dense(25, activation='relu'),
    tensorflow.keras.layers.Dense(15, activation='relu'),
    tensorflow.keras.layers.Dense(9, activation='relu')
])

model.compile(optimizer='adam', loss=tensorflow.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
# model.summary()
model.fit(X, labels, epochs=10)
```
