```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
df = pd.read_csv('original_data.csv', sep = ',')
del df["Reference"]
df.columns = ['Actuator Type','Bandwidth (Hz)',	'Strain (%)',	'Stress (MPa)',	'Efficiency (%)',	'Power Density (W/g)']
 
for (actuator), group in df.groupby(['Actuator Type']):
     group.to_csv(f'{actuator}.csv', index=False)
 
#del df["Actuator Type"]
```

```python
# Imputing Mean Values in PZT.csv
df = pd.read_csv('PZT.csv')
del df["Actuator Type"]

df.fillna(-999, inplace = True)
numpy_data_PZT = df.to_numpy()
imp = SimpleImputer(missing_values=-999, strategy='mean')
imp.fit(numpy_data_PZT)
imp = imp.transform(numpy_data_PZT)
new_column_values = ['PZT'] * 32

df = pd.DataFrame(imp, columns = [ 'Bandwidth (Hz)',	'Strain (%)',	'Stress (MPa)',	'Efficiency (%)',	'Power Density (W/g)'])
df["Actuator Type"] = new_column_values
df = df[['Actuator Type', 'Bandwidth (Hz)',	'Strain (%)',	'Stress (MPa)',	'Efficiency (%)',	'Power Density (W/g)']]
df.to_csv('PZT_mean.csv', index=False)
display(df)
```
